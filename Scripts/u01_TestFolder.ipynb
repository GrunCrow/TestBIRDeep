```python
import os
import shutil
import csv
```

```python
ROOT_PATH = "../" # "../../../desarrollo/"
```

```python
def copy_images():
    with open(ROOT_PATH + "Data/Dataset/CSVs/test.csv", "r") as csv_file:
    #with open(ROOT_PATH + "Data/Dataset/CSVs/test_with_augmented_bg.csv", "r") as csv_file:
        reader = csv.reader(csv_file)
        for row in reader:
            if row[0] != "path":
                image_path = ROOT_PATH + "Data/Dataset/images/" + row[0]
                target_dir = os.path.join(image_path.replace("images", "test").replace(os.path.basename(row[0]),""))
                target_image_path = os.path.join(target_dir, os.path.basename(row[0]))

                image_path = image_path.replace(".WAV",".PNG")
                target_image_path = target_image_path.replace(".WAV",".PNG")

                os.makedirs(target_dir, exist_ok=True)
                shutil.copy(image_path, target_image_path)
                print(f"Copied: {image_path} to {target_image_path}")

copy_images()
```

```python
def copy_audios():
    with open(ROOT_PATH + "Data/Dataset/CSVs/test_with_augmented_bg.csv", "r") as csv_file:
    #with open(ROOT_PATH + "Data/Dataset/CSVs/test_with_augmented_bg.csv", "r") as csv_file:
        reader = csv.reader(csv_file)
        for row in reader:
            if row[0] != "path":
                audio_path = ROOT_PATH + "Data/Dataset/Audios/" + row[0]
                target_dir = os.path.join(audio_path.replace("Audios", "Audios/test").replace(os.path.basename(row[0]),""))
                target_image_path = os.path.join(target_dir, os.path.basename(row[0]))

                os.makedirs(target_dir, exist_ok=True)
                shutil.copy(audio_path, target_image_path)
                print(f"Copied: {audio_path} to {target_image_path}")

copy_audios()
```
