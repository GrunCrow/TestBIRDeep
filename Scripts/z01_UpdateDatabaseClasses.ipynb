```python
# Read CSV file in Annotations folder and change specie column names No Audio to Background

import os
import pandas as pd

# Read CSV file
df = pd.read_csv('../Data/Dataset/CSVs/dataset.csv')
```

```python
df
```

```python
# Unique specie in df
df['specie'].unique()
```

```python
# len of unique specie
len(df['specie'].unique())
```

```python
# Change specie column names No Audio to Background
df['specie'] = df['specie'].replace('No audio', 'No bird')
# Sturnus to Sturnus sp.
df['specie'] = df['specie'].replace('Sturnus', 'Sturnus sp.')
# Passer to Passer sp.
df['specie'] = df['specie'].replace('Passer', 'Passer sp.')
# Lanius to Lanius sp.
df['specie'] = df['specie'].replace('Lanius', 'Lanius sp.')
# Curruca to Sylvia sp.
df['specie'] = df['specie'].replace('Curruca', 'Sylviidae')
# Sylvia to Sylviidae
df['specie'] = df['specie'].replace('Sylvia', 'Sylviidae')
# Linaria Cannabina to Linaria cannabina
df['specie'] = df['specie'].replace('Linaria Cannabina', 'Linaria cannabina')
# Galerida Cristata to Galerida cristata
df['specie'] = df['specie'].replace('Galerida Cristata', 'Galerida cristata')
# Fringilla to Fringillidae
df['specie'] = df['specie'].replace('Fringilla', 'Fringillidae')
# Sturnus to Sturnus sp.
df['specie'] = df['specie'].replace('Sturnus', 'Sturnus sp.')
```

```python
# unique specie and len
df['specie'].unique()
```

```python
# Save as dataset_updated.csv
df.to_csv('../Data/Dataset/CSVs/dataset_updated.csv', index=False)
```
